```python
dataset = Libris2sDataset(
    data_dir=".",
    split="alignments/all_de_en_alligned_cleaned.csv", 
)
print("number of samples loaded in the dataset:",len(dataset))
```

```python
# Get a sample
sample = dataset[123]

# Access the data
de_audio = sample['de_audio']  # German audio waveform
en_audio = sample['en_audio']  # English audio waveform
de_sample_rate = sample['de_sample_rate']
en_sample_rate = sample['en_sample_rate']
de_text = sample['de_transcript']  # German transcript
en_text = sample['en_transcript']  # English transcript
score = sample['alignment_score']  # Alignment quality score
```

```python
from IPython.display import display, HTML, Audio

def display_alignment_sample(sample):
    de_audio = sample['de_audio']
    en_audio = sample['en_audio']
    de_sample_rate = sample['de_sample_rate']
    en_sample_rate = sample['en_sample_rate']
    de_text = sample['de_transcript']
    en_text = sample['en_transcript']
    score = sample['alignment_score']

    html = f"""
    <h3>ðŸ”— Alignment Sample</h3>
    <p><b>Alignment Score:</b> {score:.3f}</p>
    
    <h4>German</h4>
    <p><b>Transcript:</b> {de_text}</p>
    """
    display(HTML(html))
    display(Audio(de_audio.numpy(), rate=de_sample_rate))
    
    html = f"""
    <h4>English</h4>
    <p><b>Transcript:</b> {en_text}</p>
    """
    display(HTML(html))
    display(Audio(en_audio.numpy(), rate=en_sample_rate))
```

```python
display_alignment_sample(sample)
```
